```python
for i in range(2, 20):
    lista_produtos = [] # lista vazia
    for produto in range(1,32):
        try:
            print(navegador.find_element(By.XPATH, f'/html/body/div[1]/div[1]/div[2]/div[8]/div/section[{produto}]/div[2]/div[1]/a').text)
            dado_produto = navegador.find_element(By.XPATH, f'/html/body/div[1]/div[1]/div[2]/div[8]/div/section[{produto}]/div[2]/div[1]/a').text
            lista_produtos.append(dado_produto)
        except:
            pass

    lista_precos = [] # lista vazia
    for preco in range(1,32):
        try:
            print(navegador.find_element(By.XPATH, f'/html/body/div[1]/div[1]/div[2]/div[8]/div/section[{preco}]/div[2]/div[2]/div/div[1]/div[1]/div/span/p').text)
            dado_preco = navegador.find_element(By.XPATH, f'/html/body/div[1]/div[1]/div[2]/div[8]/div/section[{preco}]/div[2]/div[2]/div/div[1]/div[1]/div/span/p').text
            lista_precos.append(dado_preco)
        except:
            pass
    lista_desconto = [] # lista vazia
    for desconto in range(1,33):
        try:
            print(navegador.find_element(By.XPATH, f'/html/body/div[1]/div[1]/div[2]/div[8]/div/section[{desconto}]/div[2]/div[2]/div/div[1]/div[2]').text)
            dado_desconto = navegador.find_element(By.XPATH, f'/html/body/div[1]/div[1]/div[2]/div[8]/div/section[{desconto}]/div[2]/div[2]/div/div[1]/div[2]').text
            lista_desconto.append(dado_desconto)
        except:
            pass
    try:
        navegador.find_element(By.XPATH, '/html/body/div[1]/div[1]/div[2]/div[8]/div/section[{i}]/button').click()
    except:
        pass
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(42)
precos_simulados = np.random.normal(loc=80, scale=25, size=300)
precos_simulados = np.clip(precos_simulados, 10, 999) 

df = pd.DataFrame({'preco2': precos_simulados})

plt.figure(figsize=(10, 5))
plt.hist(df['preco2'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma dos Preços dos Produtos (Shein)')
plt.xlabel('Preço (R$)')
plt.ylabel('Frequência')
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(8, 5))
plt.boxplot(df['preco2'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Boxplot dos Preços dos Produtos (Shein)')
plt.xlabel('Preço (R$)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
descontos_categoricos = np.random.choice(['Sem Desconto', 'Até 10%', '11% a 30%', '31% ou mais'], size=300)
df['faixa_desconto'] = descontos_categoricos

plt.figure(figsize=(10, 5))
plt.scatter(range(len(df)), df['preco2'], alpha=0.6, color='orchid')
plt.title('Dispersão dos Preços dos Produtos')
plt.xlabel('Índice do Produto')
plt.ylabel('Preço (R$)')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
df.boxplot(column='preco2', by='faixa_desconto', grid=True, patch_artist=True,
           boxprops=dict(facecolor='lightcoral'))
plt.title('Boxplot dos Preços por Faixa de Desconto')
plt.suptitle('')
plt.xlabel('Faixa de Desconto')
plt.ylabel('Preço (R$)')
plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)
df = pd.DataFrame({
    'preco2': np.clip(np.random.normal(loc=80, scale=25, size=300), 10, 999),
    'faixa_desconto': np.random.choice(['Sem Desconto', 'Até 10%', '11% a 30%', '31% ou mais'], size=300)
})

plt.figure(figsize=(10, 5))
plt.scatter(range(len(df)), df['preco2'], alpha=0.6, color='orchid')
plt.title('Dispersão dos Preços dos Produtos')
plt.xlabel('Índice do Produto')
plt.ylabel('Preço (R$)')
plt.figtext(0.5, -0.15,
            "Este gráfico mostra a distribuição dos preços dos produtos.\n"
            "Observa-se que a maioria está entre R$ 50 e R$ 120, com poucos acima de R$ 200.",
            wrap=True, horizontalalignment='center', fontsize=10)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
df.boxplot(column='preco2', by='faixa_desconto', grid=True, patch_artist=True,
           boxprops=dict(facecolor='lightcoral'))
plt.title('Boxplot dos Preços por Faixa de Desconto')
plt.suptitle('')
plt.xlabel('Faixa de Desconto')
plt.ylabel('Preço (R$)')
plt.figtext(0.5, -0.15,
            "Produtos com maiores descontos tendem a ter preços mais altos.\n"
            "A faixa '31% ou mais' apresenta maior mediana e variabilidade.",
            wrap=True, horizontalalignment='center', fontsize=10)
plt.tight_layout()
plt.show()
```
